```python
import torchaudio
from torchaudio_augmentations import Reverb
```

```python
from IPython.display import Audio
```

```python
audio1, sr1 = torchaudio.load("data/speech.wav")
Audio(audio1, rate=sr1)
audio2, sr2 = torchaudio.load("data/rock.wav")
audio2 = torchaudio.functional.resample(audio2, orig_freq=sr2, new_freq=sr1)
```

```python
reverb = Reverb(sr1)
transformed = reverb([audio1, audio2])
Audio(transformed, rate=sr1)
```
